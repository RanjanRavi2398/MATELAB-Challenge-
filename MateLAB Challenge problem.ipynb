```python
print("train shape:", train.shape)
print("Test shape:", test.shape)
df = pd.concat([train,test])
print(df.shape)
df.head()
```

```python
df['date'] = pd.to_datetime(df['DATE'],format='%Y-%m-%d')
```

```python
df['month'] = df['date'].dt.month
df['weekday'] = df['date'].dt.dayofweek
df['year'] = df['date'].dt.year
# df['date'].dt.
df['week_of_year']  = df.date.dt.weekofyear
```

```python
df.set_index("date",inplace=True)
```

```python
df["median-secondary_sales_Quanity-month"] = df.groupby(['month','Quantity',"SKU Combination","Secondary Sales Amount STC","Secondary Sales Amount Invoiced","Secondary Sales Amount Returned"])["Secondary Sales Quantity Invoiced"].transform("median")
df["mean-secondary_sales_Qunaity-week"]=df.groupby(['week_of_year','Quantity',"SKU Combination"])["Secondary Sales Quantity Invoiced"].transform("mean")
df["SKU_combination-month-sum"]=df.groupby(['month',"SKU Combination"])["Secondary Sales Quantity Invoiced"].transform("sum")
df["Quanity-month-sum"]=df.groupby(['month','Quantity'])["Secondary Sales Quantity Invoiced"].transform("mean")

```

```python
# get shifted features for grouped data. Note need to sort first! 
df['Quanity_shifted-90'] = df.groupby(["Quantity","SKU Combination"])["Secondary Sales Quantity Invoiced"].transform(lambda x:x.shift(90)) # sales for that item 90 days = 3 months ago
df['Quanity_shifted-180'] = df.groupby(["Quantity","SKU Combination"])["Secondary Sales Quantity Invoiced"].transform(lambda x:x.shift(180)) # sales for that item 180 days = 3 months ago
df['Qunaity_shifted-365'] = df.groupby(["Quantity","SKU Combination"])["Secondary Sales Quantity Invoiced"].transform(lambda x:x.shift(365)) # sales for that 1 year  ago

df["Quanity-week_shifted-90"] = df.groupby(['week_of_year','Quantity'])["Secondary Sales Quantity Invoiced"].transform(lambda x:x.shift(12).sum()) # shifted total sales for that item 12 weeks (3 months) ago
df["SKU_combination-week_shifted-90"] = df.groupby(['week_of_year',"SKU Combination"])["Secondary Sales Quantity Invoiced"].transform(lambda x:x.shift(12).sum()) # shifted total sales for that store 12 weeks (3 months) ago
df["Quanity-week_shifted-90"] = df.groupby(['week_of_year','Quantity'])["Secondary Sales Quantity Invoiced"].transform(lambda x:x.shift(12).mean()) # shifted mean sales for that item 12 weeks (3 months) ago
df["SKU_combination_shifted-90"] = df.groupby(['week_of_year',"SKU Combination"])["Secondary Sales Quantity Invoiced"].transform(lambda x:x.shift(12).mean()) # shifted mean sales for that store 12 weeks (3 months) ago
```

```python
df.head()
```

```python
col = [i for i in df.columns if i not in ['DATE','SKU Combination']]
y = 'Secondary Sales Quantity Invoiced'
```

```python
train=df.loc[~df["Secondary Sales Quantity Invoiced"].isna()]
train.shape
```

```python
test=df.loc[df["Secondary Sales Quantity Invoiced"].isna()]
test.shape
```

```python
train_x, train_cv, y, y_cv = train_test_split(train[col],train[y], test_size=0.1, random_state=42)
```

```python
def XGB_regressor(train_X, train_y, test_X, test_y, feature_names=None, seed_val=1009, num_rounds=500):
    param = {}
    param['objective'] = 'reg:linear'
    param['eta'] = 0.1
    param['max_depth'] = 6
    param['silent'] = 1
    param['eval_metric'] = 'rmse'
    param['min_child_weight'] = 1
    param['subsample'] = 0.8
    param['colsample_bytree'] = 0.8
    param['seed'] = seed_val
    num_rounds = num_rounds

    plst = list(param.items())

    xgtrain = xgb.DMatrix(train_X, label=train_y)

    if test_y is not None:
        xgtest = xgb.DMatrix(test_X, label=test_y)
        watchlist = [ (xgtrain,'train'), (xgtest, 'test') ]
        model = xgb.train(plst, xgtrain, num_rounds, watchlist, early_stopping_rounds=40)
    else:
        xgtest = xgb.DMatrix(test_X)
        model = xgb.train(plst, xgtrain, num_rounds)
        
    return model
```

```python
model = XGB_regressor(train_X = train_x, train_y = y, test_X = train_cv, test_y = y_cv)
y_test = model.predict(xgb.DMatrix(test[col]), ntree_limit = model.best_ntree_limit)
```

```python
submission['Secondary Sales Quantity Invoiced'] = y_test
submission.to_csv('submission_starter1.csv', index=False)
```
